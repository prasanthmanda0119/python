```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
iris_data=sns.load_dataset('iris')
df=sns.load_dataset('taxis')
```

```python
#to find the relationship between two variables
sns.scatterplot(x=iris_data['sepal_length'],y=iris_data['petal_length'])
plt.show()
```

```python
sns.scatterplot(x=iris_data['sepal_length'],y=iris_data['petal_length'],hue=iris_data['species'],palette=['purple','blue','black'])
plt.show()
```

```python
sns.stripplot(x=iris_data['species'],y=iris_data['sepal_width'],palette=['r','g','y'])
```

```python
sns.scatterplot(x=df['payment'],y=df['total'],palette=['purple','blue'])
plt.show()
```

```python
sns.stripplot(x=df['payment'],y=df['total'],palette=['purple','blue'])
plt.show()
```

```python
sns.stripplot(x=df['payment'],y=df['tip'],palette=['purple','blue'])
plt.show()
```

```python
sns.lineplot(x=df['tip'],y=df['fare'],color='green')
plt.show()
```

```python
sns.lineplot(x=df['total'],y=df['fare'],hue=df['payment'],palette=['green','orange'])
plt.show()
```
