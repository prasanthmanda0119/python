```python
df=sns.load_dataset("taxis")
df
```

```python
#gives the plotting
sns.barplot(x='payment',y='total',data=df,color='grey',estimator='min'
           )
plt.show()
plt.tight_layout()
```

```python
#histogram plot to know the distribution of the data
#color is used when there is a single color and palette is used when there are more than one color
sns.histplot(x=df['fare'],palette=['blue','red'],hue=df['payment'])
```

```python
sns.violinplot(x=df['fare'],y=df['payment'],palette=['green','yellow'])
plt.show()
plt.tight_layout()
```

```python
sns.boxplot(x=df['fare'],y=df['payment'],palette=['green','yellow'])
plt.show()
plt.tight_layout()
```

```python
sns.scatterplot(x=df['fare'],y=df['total'],palette=['green','yellow'],hue=df['payment'])
plt.show
plt.tight_layout()
```

```python
sns.stripplot(x=df['color'],y=df['fare'],palette=['green','yellow'])
plt.show
plt.tight_layout()
```
